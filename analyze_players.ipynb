```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import itertools
```

```python
df = pd.read_pickle('pickles/295-players.pkl')
```

```python
df
```

```python
xs = ['price', 'Rating', 'CT rating', 'T rating', 'AWP', 'HS %', 'Entry rounds', 'Clutch rounds', 'Support rounds', 'Multi kill rounds', 'Deaths per round', 'rank']

fig, axs = plt.subplots(12,1, figsize=(12, 30))
axs = axs.ravel()

plt.subplots_adjust(hspace=1)

for ax, x in zip(axs, xs):
    df.sort_values(by=x, ascending=False).plot.scatter(y=x, x='player', ax=ax)
    ax.tick_params(axis='x', rotation=90)
```

```python
fig, axs = plt.subplots(4,3, figsize=(8*4,8*3))
axs = axs.ravel()
for ax, x in zip(axs, xs):
    sns.scatterplot(data=df.sort_values('Rating'), x='player', y='price', hue=x, palette='viridis', ax=ax, zorder=3)
    ax.tick_params(axis='x', rotation=90)
    ax.margins(1e-2,1e-2)
    ax.grid(axis='x', zorder=1)
    ax.axhline(205000, color='k', zorder=2)
    ax.axhline(195000, color='k', zorder=2)
    ax.axhspan(195000, 205000, color='k', alpha=0.1, zorder=2)
```

```python
df['price/Rating'] = df.price / df.Rating
df.sort_values(by='price/Rating')[['player', 'price', 'Rating', 'price/Rating']]
```
