```python
# Lazy way

df.sort_values(by='percentage', ascending=False)
picks = pd.DataFrame(columns=df.columns)

for _ in range(5):
    tmp = df[df.apply(lambda x: x['name'] not in list(picks['name']) and x['booster'] not in list(picks['booster']), axis=1)].sort_values(by='percentage', ascending=False).iloc[0]
    picks = pd.concat([picks, tmp.to_frame().T])

display(picks.sort_values(by='name'))
print(f"{picks['percentage'].sum()=}")
```

```python
# First collect all name / role combinations that can make sense.
# Keep looking for another role for a player, if at least one more 
# player has that role as maximum role also.

df_role_per_name = df.groupby('name', group_keys=True).apply(lambda x: x[['booster', 'percentage']].sort_values(by='percentage', ascending=False, ignore_index=True))

names = df['name'].unique()
roles = range(len(df['booster'].unique()))

options_dict = {}
roleidx = 0
for roleidx in roles:
    for name in names:
        role, percentage = df_role_per_name.loc[name].iloc[roleidx]

        try:
            options_dict[role]['names'].append({'name': name, 'percentage': percentage})
            options_dict[role]['need_another_pass'] = True
        except KeyError:
            options_dict[role] = {'names': [{'name': name, 'percentage': percentage}], 'need_another_pass': False}

    names = []

    for role in options_dict.keys():
        if options_dict[role]['need_another_pass'] and len(options_dict[role]['names']) > 1:
            # print(f'{role=} needs another pass')
            names = [x['name'] for x in options_dict[role]['names']]
        options_dict[role]['need_another_pass'] = False

    if not names:
        break

options_tuple = []
for role in options_dict.keys():
    for name in options_dict[role]['names']:
        options_tuple.append((role, name['name'], name['percentage']))

# Generate all 5 pick options, then filter out proper ones.
combs = itertools.combinations(options_tuple, r=5)
def filter_combs(combs):
    for comb in combs:
        roles, names, _ = zip(*comb)
        if len(set(roles)) == len(roles) and len(set(names)) == len(names):
            yield comb
filtered_combs = filter_combs(combs)

# Find maximum sum.
def sum_comb(comb):
    _,_,perc = zip(*comb); 
    return sum(perc)
max_comb = max(filtered_combs, key=lambda x: sum_comb(x))

display(pd.DataFrame(max_comb, columns=['role', 'name', 'percentage']).sort_values(by='name'))
print(f'{sum_comb(max_comb)=}')
```
